# K-Means Initialization

A short clustering exercise showing how different starting centers can lead K-Means through different optimization paths.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate a small synthetic dataset with three visible clusters.
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=42)

# Try two different starting points for the cluster centers.
init1 = np.array([[0, 0], [1, 1], [2, 2]])
init2 = np.array([[4, 4], [-4, -4], [0, 0]])

# Fit K-Means from each initialization.
kmeans1 = KMeans(n_clusters=3, init=init1, n_init=1, max_iter=300, random_state=1)
kmeans2 = KMeans(n_clusters=3, init=init2, n_init=1, max_iter=300, random_state=1)

labels1 = kmeans1.fit_predict(X)
labels2 = kmeans2.fit_predict(X)

# Compare the final assignments side by side.
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].scatter(X[:, 0], X[:, 1], c=labels1, cmap='viridis', s=30)
axs[0].scatter(kmeans1.cluster_centers_[:, 0], kmeans1.cluster_centers_[:, 1], c='red', marker='X', s=200)
axs[0].set_title("Initialization 1")

axs[1].scatter(X[:, 0], X[:, 1], c=labels2, cmap='plasma', s=30)
axs[1].scatter(kmeans2.cluster_centers_[:, 0], kmeans2.cluster_centers_[:, 1], c='black', marker='X', s=200)
axs[1].set_title("Initialization 2")

plt.suptitle("Effect of Different Initializations on K-Means Clustering")
plt.show()
```
